```python
#Q3. What is App routing in Flask? Why do we use app routes?
```

```python
""" In Flask, App routing refers to the process of mapping the URLs of a web application to Python functions that will handle the requests
 made to those URLs. This is done using decorators in Flask.
 
 We use app routes in Flask to create different URLs for different pages or functionality in a web application. By mapping URLs to specific
     Python functions, we can easily handle user requests and return the appropriate responses.

Using app routes also makes it easy to organize and maintain a Flask application, as each route can be mapped to a specific function, making
    the code easier to read and understand. Additionally, app routes make it easier to modify the behavior of a web application without having
    to change the URL structure."""
```

```python
#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
```

```python
""" 
In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint with the values provided as arguments. 
It takes the endpoint name as the first argument and any number of keyword arguments representing the variable parts of the URL. The url_for()
function generates a URL that corresponds to the endpoint and the variable parts provided.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return 'Post %d' % post_id

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index')) # output: /
        print(url_for('show_user_profile', username='Awinash kumar')) # output: /user/Awinash kumar
        print(url_for('show_post', post_id=1)) # output: /post/1




"""
```
